```python
import pandas as pd
import datetime
import numpy as np
import math
import re
from matplotlib import pyplot as plt

def formatStockPrices(stock_price):
    x = re.sub('\.',"",stock_price)
    x = re.sub(',', '.', x)
    return x

def shift(df, t):
    for _ in range(t):
        df['open'] = df['open'].shift(-1)
        df.drop(df.tail(1).index,inplace=True)

coffee_prices = pd.read_csv('data/coffee-prices.csv')
coffee_prices["date"] = pd.to_datetime(coffee_prices['date'])
mask = (coffee_prices['date'] >= datetime.datetime(2019,1,1)) & (coffee_prices['date'] <= datetime.datetime(2019,12,31))
coffee_prices = coffee_prices.loc[mask]

oil_prices = pd.read_csv('data/oil-prices.csv')
oil_prices['date'] = pd.to_datetime(oil_prices['date'])
mask = (oil_prices['date'] >= datetime.datetime(2019,1,1)) & (oil_prices['date'] <= datetime.datetime(2019,12,31))
oil_prices = oil_prices.loc[mask]

stock_prices = pd.read_csv('data/colcap-prices.csv')[['Fecha','ultimo']]
stock_prices['Fecha'] = pd.to_datetime(stock_prices['Fecha'])
stock_prices['ultimo'] = stock_prices['ultimo'].apply(formatStockPrices)
stock_prices['ultimo'] = pd.to_numeric(stock_prices['ultimo'])
stock_prices.columns = ['date','close']
stock_prices = stock_prices.sort_values('date')

exchange_rates = pd.read_csv("data/exchange-rates.csv")[["Date","Open"]]
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])
exchange_rates.columns = ['date','open']
#shift(exchange_rates, 5)

print(coffee_prices)
print(oil_prices)
print(stock_prices)
print(exchange_rates)
```

```python
coffee_to_exchange = pd.merge(coffee_prices,exchange_rates,'inner','date')
coffee_to_exchange = coffee_to_exchange.drop(columns=['date'])

oil_to_exchange = pd.merge(oil_prices, exchange_rates, 'inner', 'date')
oil_to_exchange = oil_to_exchange.drop(columns=['date'])

stock_to_exchange = pd.merge(stock_prices, exchange_rates, 'inner', 'date')
stock_to_exchange = stock_to_exchange.drop(columns=['date'])

coffee_to_oil = pd.merge(coffee_prices,oil_prices,'inner','date')
coffee_to_oil = coffee_to_oil.drop(columns=['date'])

print(coffee_to_exchange)
print(oil_to_exchange)
print(stock_to_exchange)
print(coffee_to_oil)
```

```python
r_ce = coffee_to_exchange.corr(method='pearson')
r_ce = r_ce.to_numpy()[0][1]

r_oe = oil_to_exchange.corr(method='pearson')
r_oe = r_oe.to_numpy()[0][1]

r_se = stock_to_exchange.corr(method='pearson')
r_se = r_se.to_numpy()[0][1]

r_co = coffee_to_oil.corr(method='pearson')
r_co = r_co.to_numpy()[0][1]

print(r_ce)
print(r_oe)
print(r_se)
print(r_co)
```

```python
def plot(df):
    df.plot.scatter(x=df.columns[0],y=df.columns[1])
plot(coffee_to_exchange)
plot(oil_to_exchange)
plot(stock_to_exchange)
```

```python
def addToGraph(dfs):
    plt.figure(211)
    for i, df in enumerate(dfs):
        x, y = df.columns
        plt.subplot(int('21'+str(i+1)))
        plt.plot(df[x],df[y])
    plt.show()
addToGraph([stock_prices, exchange_rates])
```
